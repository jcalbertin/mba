## Install

```r
install.packages("nnfor")
```

### The nnfor (development version here) package for R facilitates time series forecasting with Multilayer Perceptrons (MLP) and Extreme Learning Machines (ELM)

```r
library(nnfor)
fit.ts <- mlp(AirPassengers, hd=c(5,4))
print(fit.ts)
```

```r
plot(fit.ts)
```

```r
frc <- forecast(fit.ts,h=12)
print(frc)
```

```r
plot(frc)
```

### ELM

```r
fit.elm.ts <- elm(AirPassengers)
print(fit.elm.ts)
```

```r
plot(fit.elm.ts)
```

```r
frc <- forecast(fit.elm.ts,h=12)
print(frc)
```

```r
plot(frc)
```
